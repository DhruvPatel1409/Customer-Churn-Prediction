```python
X = df.drop(['Exited'],axis=1)
Y = df['Exited']
```

```python
plt.figure(figsize=(15,7))
sns.countplot(x=Y)
```

```python
from imblearn.over_sampling import SMOTE # synthetic minority oversampling technique
smote = SMOTE()
X,Y=smote.fit_resample(X,Y)
```

```python
plt.figure(figsize=(15,10))
sns.countplot(x=Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(random_state=42)
rfc.fit(X_train,Y_train)
```

```python
y_pred = rfc.predict(X_test)
print(y_pred)
```

```python
print("Accuracy_score :", round(accuracy_score(Y_test,y_pred)*100,2),"%")
```

```python
from sklearn.metrics import classification_report
print(classification_report(Y_test,y_pred))
```

```python
from sklearn.metrics import confusion_matrix
print(confusion_matrix(Y_test,y_pred))
```

```python
sns.heatmap(confusion_matrix(Y_test,y_pred),annot=True,fmt='d', xticklabels=['Predicted 0', 'Predicted 1'],
            yticklabels=['Actual 0', 'Actual 1'])
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.title('CONFUSION MATRIX FOR RANDOM FOREST CLASSIFIER')
```

```python
from sklearn.metrics import roc_curve, roc_auc_score
Y_pred_pro = rfc.predict_proba(X_test)[:][:,1]
df_actual_predicted = pd.concat([pd.DataFrame(np.array(Y_test), columns=['Y_actual']), pd.DataFrame(Y_pred_pro, columns=['Y_pred_pro'])], axis=1)
df_actual_predicted.index = Y_test.index
fpr, tpr, tr = roc_curve(df_actual_predicted['Y_actual'], df_actual_predicted['Y_pred_pro'])
auc = roc_auc_score(df_actual_predicted['Y_actual'], df_actual_predicted['Y_pred_pro'])
plt.plot(fpr, tpr, label='AUC = %0.4f' %auc)
plt.plot(fpr, fpr, linestyle = '--', color='k')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve', size = 15)
plt.legend()
```
